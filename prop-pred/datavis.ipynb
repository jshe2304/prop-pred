```python
import pandas as pd

import matplotlib.pyplot as plt
```

# Immunomodulation Dataset Visualizations

```python
data = pd.concat((pd.read_csv('test.csv'), pd.read_csv('train.csv')))

N, n_columns = data.shape
properties = [col for col in data.columns[3:] if 'Std' not in col]
print(f'Number of Molecules = {N}')
print(f'Number of Columns = {n_columns}')
print(f'Number of Properties = {len(properties)}')
```

```python
def plot_property(ax, data, property_label):
    ax.hist(data, bins=32)
    ax.set_title(f'Distribution of {property_label}')
    ax.set_xlabel(property_label)
    ax.set_ylabel('Counts')

    #annotation = 'Range = [{:.2f}, {:.2f}]\n'.format(data.min(), data.max()) 
    annotation = 'Med = {:.2f}\n'.format(data.median()) 
    annotation += '$\\mu$ = {:.2f}\n'.format(data.mean()) 
    annotation += '$\\sigma$ = {:.2f}\n'.format(data.std()) 
    ax.text(1.1, 0.9, annotation, transform=ax.transAxes , fontsize=10, verticalalignment='top')
```

```python
fig, axs = plt.subplots(len(properties), 1, figsize=(6, 4 * len(properties)))

for ax, label in zip(axs, properties):
    plot_property(ax, data[label], label)

plt.tight_layout()
```

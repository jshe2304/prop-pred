```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_bnn_emb = pd.read_csv('./bnn_embeddings.csv')
data_bnn_fp = pd.read_csv('./bnn_fingerprints.csv')
data_mlp_emb = pd.read_csv('../../mlp/logs/distributions/rmlp_emb_multi_1.csv')
data_mlp_fp = pd.read_csv('../../mlp/logs/distributions/rmlp_fp_multi_1.csv')
```

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 6))
axs = axs.flatten()

for i, property_label in enumerate(data_bnn_emb.columns):

    ax = axs[i]

    ax.boxplot(
        x=[data_bnn_emb.iloc[:, i], data_bnn_fp.iloc[:, i], data_mlp_emb.iloc[:, i], data_mlp_fp.iloc[:, i]], 
        tick_labels=['BNN (Emb)', 'BNN (FP)', 'MLP (Emb)', 'MLP (FP)'], 
        vert=False, sym='+', 
        medianprops={'color': 'black'}
    )

    ax.set_xlabel(r'$R^2$')
    ax.set_title(property_label)

axs[-1].axis('off')
axs[-2].axis('off')

#fig.legend(handles=legend_elements, bbox_to_anchor=(0.5, 0.125), loc='lower center')
fig.suptitle('Performance of Bayesian prediction head', fontsize=18)

plt.tight_layout()
```

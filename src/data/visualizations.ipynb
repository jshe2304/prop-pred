```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
split = 'train'
property_name = 'LPS_Abs'

column_name = 'property_' + property_name

log_standardize_prefix = pd.read_csv('./log_standardize_prefix/' + split + '.csv')
standardize_prefix = pd.read_csv('./standardize_prefix/' + split + '.csv')
prefix = pd.read_csv('./prefix/' + split + '.csv')

fig, ((ax00, ax01, ax02), (ax10, ax11, ax12)) = plt.subplots(2, 3, figsize=(9, 6))

ax00.boxplot(prefix[column_name], vert=False, whis=3, sym='.', flierprops={'markersize': 4})
ax00.set_xlabel('$x$')
ax00.set_yticks([])

ax01.boxplot(standardize_prefix[column_name], vert=False, whis=4, sym='.', flierprops={'markersize': 4})
ax01.set_xlabel('$Standardize(x)$')
ax01.set_yticks([])

ax02.boxplot(log_standardize_prefix[column_name], vert=False, whis=4, sym='.', flierprops={'markersize': 4})
ax02.set_xlabel('$Standardize \\circ \\log(x)$')
ax02.set_yticks([])

ax10.scatter(prefix[column_name], standardize_prefix[column_name], s=1, alpha=0.25)
ax10.set_xlabel('$x$')
ax10.set_ylabel('$Standardize(x)$')

ax11.scatter(prefix[column_name], log_standardize_prefix[column_name], s=1, alpha=0.25)
ax11.set_xlabel('$x$')
ax11.set_ylabel('$Standardize \\circ \\log(x)$')

ax12.scatter(standardize_prefix[column_name], log_standardize_prefix[column_name], s=1, alpha=0.25)
ax12.set_xlabel('$Standardize(x)$')
ax12.set_ylabel('$Standardize \\circ \\log(x)$')

fig.suptitle(f'Data and Transforms for $x=${property_name} ')

plt.tight_layout()
```

```python
fig.savefig(f'./figures/{property_name}_{split}.pdf')
```
